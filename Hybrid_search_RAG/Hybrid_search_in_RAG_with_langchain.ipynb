```python
from google.colab import userdata
pinecode_api_key=userdata.get('pinecode_api_key')
hf_token=userdata.get('huggingface_token')
```

```python
from langchain_community.retrievers import PineconeHybridSearchRetriever
```

```python
import os
from pinecone import Pinecone,ServerlessSpec
```

```python
index_name="hybrid-search-demo"
```

```python
#initialize pinecone client
pc=Pinecone(api_key=pinecode_api_key)

if index_name not in pc.list_indexes().names():
  pc.create_index(
      name=index_name,
      dimension=384,
      metric="dotproduct",
      spec=ServerlessSpec(cloud="aws",region="us-east-1"),
  )
```

```python
# !pip install --upgrade transformers
```

```python
index=pc.Index(index_name)
index
```

```python
from langchain.embeddings import HuggingFaceEmbeddings
embeddings=HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
embeddings
```

```python
from pinecone_text.sparse import BM25Encoder
```

```python
bm25_encoder=BM25Encoder().default()
bm25_encoder
```

```python
sentences=[
    "In 2023, I visited Paris",
    "In 2022, I visited New York",
    "In 2021, I visited London",
    "In 2020, I visited Tokyo",
    "In 2019, I visited Sydney",
]
```

```python
bm25_encoder.fit(sentences)
```

```python
bm25_encoder.dump("bm25_encoder.json")
```

```python
bm25_encoder=BM25Encoder().load("bm25_encoder.json")
bm25_encoder
```

```python
retrievers=PineconeHybridSearchRetriever(
    embeddings=embeddings,
    sparse_encoder=bm25_encoder,
    index=index,
)
```

```python
retrievers
```

```python
retrievers.add_texts([
    "In 2023, I visited Paris",
    "In 2022, I visited New York",
    "In 2021, I visited London",
    "In 2020, I visited Tokyo",
    "In 2019, I visited Sydney",
])
```

```python
retrievers.invoke("What year was Paris visited?")
```
